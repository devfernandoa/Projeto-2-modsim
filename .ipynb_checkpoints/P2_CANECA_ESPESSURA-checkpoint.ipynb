```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pandas
```

```python
#Variaveis estaticas

#Temperaturas do ambiente, recipiente e do cafe, respectivamente, em K
Ta = 24 + 273.15
Tc = 96 + 273.15
Tr = 24 + 273.15

#Dimens√µes da caneca, r (raio interno), R(raio externo), rc(raio intermediario) e h(altura) em metros:
r = 0.04
R = 0.043
re = (r + R) / 2
h = 0.1

#Massa do cafe e do recipiente em kg, respectivamente:
mc = 0.25
mr = 0.1

#Calores especificos do cafe e do recipiente, em J/(kg*K)
cc = 4186
cr = 840

#Condutividade termica do vidro em W/m‚ãÖùêæ
K = 0.8

#coeficiente de convec√ß√£o do ar em ùëä/(ùëö^2‚ãÖùêæ)
hs = 24
```

```python
def modelo(l, t):
    Tc = l[0]
    Tr = l[1]
    Ti = Tc
    
    qs = (Tc - Ta) / (1 / (hs * math.pi * r ** 2))
    qr = (Tc - Tr) / ((math.log(re / r))/ (K * 2 * math.pi * h))
    qf = (Ti - Ta) / ((math.log(R/re) / (K * 2 * math.pi * h)) + (1 / (hs * 2 * math.pi * R * h)))
    dTcdt = (-1 / (mc * cc)) * (qs + qr)
    dTrdt = (1 / (mr * cr)) * (qr - qf)
    return [dTcdt, dTrdt]
```

```python
lista_tempo = np.arange(0, 5000, 2)
ci = [Tc, Tr]

x = odeint(modelo, ci, lista_tempo)
tCafe = x[:,0]
tRecipiente = x[:,1]

plt.plot(lista_tempo/3600, tCafe - 273.15, label='Temperatura do Caf√©')
plt.plot(lista_tempo/3600, tRecipiente - 273.15, label='Temperatura do recipiente')
plt.legend()
plt.xlabel("Tempo (segundos)")
plt.ylabel("Temperatura do Cafe (C)")
plt.grid(True)
plt.show()
```

```python
def modelo2(l, t, R):
    mr = ((math.pi * (R ** 2) * h) - (math.pi * (r ** 2) * (h-0.01))) * 2500
    print(mr)
    re = (r + R) / 2
    
    Tc = l[0]
    Tr = l[1]
    Ti = Tr
    
    qs = (Tc - Ta) / (1 / (hs * math.pi * r ** 2))
    qr = (Tc - Tr) / ((math.log(re / r))/ (K * 2 * math.pi * h))
    qf = (Ti - Ta) / ((math.log(R/re) / (K * 2 * math.pi * h)) + (1 / (hs * 2 * math.pi * R * h)))
    dTcdt = (1 / (mc * cc)) * (-qs - qr)
    dTrdt = (1 / (mr * cr)) * (qr - qf)
    return [dTcdt, dTrdt]
```

```python
lista_espessura = np.arange(0, 0.01, 0.001)
lista_odeint = []

for i in lista_espessura:
    x = odeint(modelo2, ci, lista_tempo, args=(R+i,))
    lista_odeint.append(x[:,0])
    plt.plot(lista_tempo / 3600, x[:,0] - 273.15, label = i)


plt.legend()
plt.xlabel("Tempo (horas)")
plt.ylabel("Temperatura do Cafe (C)")
plt.grid(True)
plt.show()
```

```python
def conclusivo(lo, lt):
    temp = 60 + 273.15
    
    lista_tcafe = list(lo)
    lista_tempo = list(lt)
    index = int()
    
    for i in lista_tcafe:
        if i <= temp:
            index = lista_tcafe.index(i)
            break
    return lista_tempo[index]

for i in lista_odeint:
    print(conclusivo(i, lista_tempo))
```
