```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import pandas
```

```python
#Variaveis estaticas
#RECIPIENTE DE CERAMICA

#Temperaturas do ambiente, cafe e do recipiente, respectivamente, em K
Ta = 25.6 + 273.15
Tc = 75.20 + 273.15
Tr = 67.3 + 273.15

#Dimens√µes da caneca, r (raio interno), R(raio externo), rc(raio intermediario) e h(altura) em metros:
r = 0.04 - 0.003
R = 0.04
rc = (r + R) / 2
h = 0.09

#Massa do cafe e do recipiente em kg, respectivamente:
mc = (math.pi * (r ** 2) * (0.047)) * 1000
mr = ((math.pi * (R ** 2) * h) - (math.pi * (r ** 2) * (h-0.01))) * 2480

#Calores especificos do cafe e do recipiente, em J/(kg*K)
cc = 4186
cr = 1070

#Condutividade termica da ceramica em W/m‚ãÖùêæ
K = 1.5

#coeficiente de convec√ß√£o do ar em ùëä/(ùëö^2‚ãÖùêæ)
hs = 42
```

```python
def modelo(l, t):
    Tc = l[0]
    Tr = l[1]
    Ti = Tc
    
    qs = (Tc - Ta) / (1 / (hs * math.pi * r ** 2))
    qr = (Tc - Tr) / ((math.log(rc / r))/ (K * 2 * math.pi * h))
    qf = (Ti - Ta) / ((math.log(R/rc) / (K * 2 * math.pi * h)) + (1 / (hs * 2 * math.pi * R * h)))
    dTcdt = (-1 / (mc * cc)) * (qs + qr)
    dTrdt = (1 / (mr * cr)) * (qr - qf)
    return [dTcdt, dTrdt]
```

```python
lista_tempo = np.arange(0, 1500, 2)
ci = [Tc, Tr]

x = odeint(modelo, ci, lista_tempo)
tCafe = x[:,0] - 273.15
tRecipiente = x[:,1] - 273.15

plt.plot(lista_tempo/3600, tCafe, label='Temperatura do Caf√©')
plt.plot(lista_tempo/3600, tRecipiente, label='Temperatura do recipiente')
plt.legend()
plt.xlabel("Tempo (horas)")
plt.ylabel("Temperatura do Cafe (C)")
plt.grid(True)
plt.show()
```

```python
def conclusivo(lo, lt):
    temp = 60 + 273.15
    
    lista_tcafe = list(lo)
    lista_tempo = list(lt)
    index = int()
    
    for i in lista_tcafe:
        if i <= temp:
            index = lista_tcafe.index(i)
            break
    return lista_tempo[index]
```
